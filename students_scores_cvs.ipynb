```python
# Student Score Analyzer
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("students_scores.csv")

# Show the first 5 rows
print("ðŸ“‹ First 5 rows of data:")
print(df.head())

# Basic stats
average_score = df["Score"].mean()
highest_score = df["Score"].max()
lowest_score = df["Score"].min()

print("\nðŸ“Š Insights:")
print(f"Average Score: {average_score:.2f}")
print(f"Highest Score: {highest_score}")
print(f"Lowest Score: {lowest_score}")

# Pass/Fail count (pass mark = 50)
pass_mark = 50
passed = (df["Score"] >= pass_mark).sum()
failed = (df["Score"] < pass_mark).sum()

print(f"Number of Students Passed: {passed}")
print(f"Number of Students Failed: {failed}")

# Plot scores
plt.figure(figsize=(8,5))
plt.bar(df["Name"], df["Score"], color="skyblue")
plt.axhline(y=50, color="r", linestyle="--", label="Pass Mark")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores Analysis")
plt.legend()
plt.show()
```
